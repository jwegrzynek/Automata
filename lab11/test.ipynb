# Test FFM dla różnych parametrów

```python
import numpy as np
import matplotlib.pyplot as plt
from ex18 import generate_random_grid, update_grid


def run_simulation(TREE_PROPORTION, GROW_TREE_PROB, FIRE_PROB, FIRE_PROB_WIND_DIFF_SIDE, WIND_DIRECTION, steps=250):
    cell_size = 4
    res = (800, 800)
    grid = generate_random_grid(int(res[0] / cell_size), TREE_PROPORTION)

    state_counts = {
        'empty': [],
        'tree': [],
        'burning': []
    }

    for _ in range(steps):
        grid = update_grid(grid, FIRE_PROB, GROW_TREE_PROB, FIRE_PROB_WIND_DIFF_SIDE, WIND_DIRECTION)

        # Count states
        unique, counts = np.unique(grid, return_counts=True)
        count_dict = dict(zip(unique, counts))
        state_counts['empty'].append(count_dict.get(0, 0))
        state_counts['tree'].append(count_dict.get(1, 0))
        state_counts['burning'].append(count_dict.get(2, 0))

    time_steps = range(len(state_counts['empty']))
    plt.figure(figsize=(10, 6))
    plt.plot(time_steps, state_counts['empty'], label='Empty', color='black')
    plt.plot(time_steps, state_counts['tree'], label='Tree', color='green')
    plt.plot(time_steps, state_counts['burning'], label='Burning', color='red')
    plt.xlabel('Time step')
    plt.ylabel('Count')
    plt.title(f'Forest Fire Model: Cell State Counts Over Time')
    plt.legend()
    plt.show()
```

Pan testował model bez wiatru to ja przeanalizuję ten z wiatrem (po porównaniu kilku modeli z wiatrem i bez, można powiedzieć, że wpływa on głównie na to na jakim poziomie ustabilizuje się populacja drzew). Kierunek wiatru jest w sumie bez znaczenia. Fajnie wygląda w samej wizualizacji :)
